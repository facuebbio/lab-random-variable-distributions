```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## Loading the dataset into the variable "customer_df"
file_path = "files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv"
customer_df = pd.read_csv(file_path)
customer_df.head()
```

```python
## Getting numerical variables from out dataset
numerical_variables = customer_df.select_dtypes(include=['int64', 'float64']).columns.tolist()
numerical_variables
```

```python
## Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.

fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(14, 22))
fig.tight_layout(pad=5.0)

for i, var in enumerate(numerical_variables):
    sns.histplot(customer_df[var], kde=True, ax=axes[i//2, i%2], color="green").set_title(var)
plt.show()
```

```python
## Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.

variables_to_transform = ['Customer Lifetime Value', 'Monthly Premium Auto', 'Total Claim Amount']
customer_df_log = customer_df.copy()

for var in variables_to_transform:
    customer_df_log[var] = np.log(customer_df_log[var])

# Creating distribution plots for the log-transformed variables
fig, axes = plt.subplots(nrows=len(variables_to_transform), ncols=1, figsize=(12, 16))
fig.tight_layout(pad=5.0)

for i, var in enumerate(variables_to_transform):
    sns.histplot(customer_df_log[var], kde=True, ax=axes[i], color="blue").set_title(f'Log-transformed {var}')

plt.show()
```

- **Customer Lifetime Value**: The log transformation enhances the distribution's bell shape, suggesting a closer alignment with normal distribution assumptions post-transformation.

- **Monthly Premium Auto**: Similarly, the data becomes more normally distributed, with a notable reduction in skewness, highlighting the effectiveness of log transformation in addressing right-skewed data.

- **Total Claim Amount**: Additionally, there is progress towards a normal distribution, although slight skewness persists. Nevertheless, the transformation aids in achieving greater symmetry in the data distribution.
